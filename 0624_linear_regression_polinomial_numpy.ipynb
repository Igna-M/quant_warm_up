```python
X = np.random.rand(100,1)
y = 4 + 5*X + np.random.rand(100,1)
```

```python
plt.figure(figsize=(12,9))
# plt.scatter(X,y)
# sns.scatterplot(data=df, x='X', y='y')
sns.scatterplot(x=X[:,0], y=y[:,0])
plt.axhline(y.mean(), color='green', linewidth=1, label='Media')
plt.legend()
plt.show();
```

```python
fig, ax = plt.subplots(figsize=(18, 6))
sns.histplot(x=X[:,0], bins='auto', ax=ax)
sns.histplot(x=y[:,0], bins=20, ax=ax, color='orange')
plt.show()
```

```python
reg = LinearRegression()
```

```python
reg.fit(X, y)
```

```python
X_vals = np.linspace(0,1,100).reshape(-1, 1)
y_vals = reg.predict(X_vals)
```

```python
plt.figure(figsize=(18,9))
plt.grid(True)
sns.scatterplot(x=X[:,0], y=y[:,0])
sns.regplot(x=X[:,0], y=y[:,0])
sns.lineplot(x=X_vals[:,0], y=y_vals[:,0], color='r', linewidth=2, label='Regresión lineal')
plt.axhline(y.mean(), color='green', linewidth=1, label='Media')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresión Lineal')
plt.legend()
plt.show();
```

```python
y_test = y
y_pred = reg.predict(X)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R2 score:", r2)
```

```python
reg.score(X, y)
```

```python
beta = reg.coef_
intercept = reg.intercept_

beta, intercept
```

```python
X_sm = sm.add_constant(X)
model = sm.OLS(y, X_sm).fit()
```

```python
print(model.summary())
```

```python
predictions = model.predict(X_sm)
```

```python
plt.figure(figsize=(18,9))
plt.grid(True)
#sns.scatterplot(data=df, x='X', y='y')
sns.regplot(x=X[:,0], y=y[:,0])
sns.lineplot(x=X_vals[:,0], y=y_vals[:,0], color='r', linewidth=2, label='Regresión lineal scikit learn')
plt.plot(X, predictions, color='purple', linewidth=1, label='Regresión lineal statsmodels')
plt.axhline(y.mean(), color='green', linewidth=1, label='Media')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresión Lineal')
plt.legend()
plt.show();
```

```python
X = 4 * np.random.rand(100,1) - 2
y = 4 + 2*X + 5*X**2 + np.random.rand(100,1)
```

```python
plt.figure(figsize=(12,9))
sns.scatterplot(x=X[:,0], y=y[:,0])
plt.axhline(y.mean(), color='green', linewidth=1, label='Media')
plt.legend()
plt.show();
```

```python
fig, ax = plt.subplots(figsize=(18, 6))
sns.histplot(x=X[:,0], bins='auto', ax=ax)
sns.histplot(x=y[:,0], bins=20, ax=ax, color='orange')
plt.show()
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
poly
```

```python
X_poly = poly.fit_transform(X)
```

```python
X_poly[:5]
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_poly, y)
```

```python
y_test = y
y_pred = reg.predict(X_poly)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R2 score:", r2)
```

```python
reg.score(X_poly, y)
```

```python
beta = reg.coef_
intercept = reg.intercept_

beta, intercept
```

```python
X_sm = sm.add_constant(X_poly)
sm_model = sm.OLS(y, X_sm).fit()
```

```python
predictions = sm_model.predict(X_sm)
```

```python
print(sm_model.summary())
```

```python
X_linespace = np.linspace(-2,2,100).reshape(-1, 1)
X_linespace_poly = poly.transform(X_linespace)
X_linespace_poly_pred = reg.predict(X_linespace_poly)
```

```python
plt.figure(figsize=(18,9))
plt.grid(True)
sns.scatterplot(x=X[:,0], y=y[:,0])
sns.regplot(x=X_poly[:,0], y=y[:,0])
sns.lineplot(x=X_linespace[:,0], y=X_linespace_poly_pred[:,0], color='r', linewidth=2, label='Regresión lineal scikit learn')
sns.lineplot(x=X_sm[:,1], y=predictions, color='purple', linewidth=2, label='Regresión lineal statsmodels')
plt.axhline(y.mean(), color='green', linewidth=1, label='Media')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresión Lineal Polinomial')
plt.legend()
plt.show();
```
